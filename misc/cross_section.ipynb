Plotting neutrino cross sections in lead, from "What Can Be Learned with a Lead-Based Supernova-Neutrino Detector?" by Engel, McLaughlin, and Volpe

https://arxiv.org/pdf/hep-ph/0209267.pdf

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
plt.style.use('jfc') # my custom style sheet
%config InlineBackend.figure_format = 'retina'
```

```python
E = np.arange(5,100,5) # energies in MeV

CC1n = np.array([0.29e-11,0.91,4.96,14.66,25.05,29.27,33.56,37.91,42.54,47.17,52.02,56.31,60.39,64.03,67.04,69.69,71.95,73.91])
CC2n = np.array([0,0.45,3.15,10.85,23.68,38.97,53.79,71.63,90.05,108.73,129.14,150.4,170.75,191.16,211.73,231.25])

n1n = np.array([0,0.002,0.06,0.20,0.46,0.87,1.44,2.15,2.97,3.86,4.79,5.74,6.71,7.69,8.67,9.65,10.58,11.45,12.23])
n2n = np.array([0,0.03,0.15,0.42,0.93,1.74,2.93,4.56,6.63,9.17,12.17,15.59,19.39,23.51,27.90,32.47])

nb1n = np.array([0,0.002,0.05,0.18,0.40,0.73,1.18,1.73,2.34,2.99,3.65,4.31,4.97,5.62,6.25,6.86,7.44,7.97,8.45])
nb2n = np.array([0,0.03,0.13,0.36,0.76,1.39,2.26,3.42,4.85,6.54,8.47,10.62,12.94,15.39,17.93,20.51])
```

```python
xCC1n = np.linspace(10,95,1000)
fCC1n = interp1d(E[1:], CC1n, kind='cubic')

xCC2n = np.linspace(20,95,1000)
fCC2n = interp1d(E[3:], CC2n, kind='cubic')

xn1n = np.linspace(5,95,1000)
fn1n = interp1d(E, n1n, kind='cubic')

xn2n = np.linspace(20,95,1000)
fn2n = interp1d(E[3:], n2n, kind='cubic')

xnb1n = np.linspace(5,95,1000)
fnb1n = interp1d(E, nb1n, kind='cubic')

xnb2n = np.linspace(20,95,1000)
fnb2n = interp1d(E[3:], nb2n, kind='cubic')
```

```python
fig,ax = plt.subplots(1,1)

ax.plot(xCC1n,fCC1n(xCC1n),c='C1',label=r'$\nu_e \rightarrow 1n$',zorder=1)
ax.plot(xCC2n,fCC2n(xCC2n),c='C1',linestyle=':',label=r'$\nu_e \rightarrow 2n$',zorder=2)
#ax.vlines(xCC2n[0],0,fCC2n(xCC2n)[0],colors='C1',linestyle=':',zorder=2)

ax.plot(xn1n,fn1n(xn1n),c='C0',label=r'NC$\ \nu \rightarrow 1n$',zorder=3)
#ax.vlines(xn1n[0],0,fn1n(xn1n)[0],colors='C0',zorder=3)
ax.plot(xn2n,fn2n(xn2n),c='C0',linestyle=':',label=r'NC$\ \nu \rightarrow 2n$',zorder=4)
#ax.vlines(xn2n[0],0,fn2n(xn2n)[0],colors='C0',linestyle=':',zorder=4)

ax.plot(xnb1n,fnb1n(xnb1n),c='C2',label=r'NC$\ \bar{\nu} \rightarrow 1n$',zorder=5)
#ax.vlines(xnb1n[0],0,fnb1n(xnb1n)[0],colors='C2',zorder=5)
ax.plot(xnb2n,fnb2n(xnb2n),c='C2',linestyle=':',label=r'NC$\ \bar{\nu} \rightarrow 2n$',zorder=6)
#ax.vlines(xnb2n[0],0,fnb2n(xnb2n)[0],colors='C2',linestyle=':',zorder=6)

ax.legend()

ax.set_xlabel("Neutrino Energy (MeV)")
ax.set_ylabel("Cross section ($10^{-40}$ cm$^2$)")

ax.set_yscale('log')

ax.set_xlim(0,90)
ax.set_ylim(1e-6,1e3)

fig.savefig("cross_sections.pdf")
plt.show()
```
