## Figure of merit plot for HALO 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
figsize = (8,6)
plt.rc('font', family='sans-serif')
plt.rc('axes', labelsize='x-large')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
plt.rc('axes',titlesize='x-large')
#plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['savefig.dpi'] = 200
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams['savefig.pad_inches'] = 0.1
#print(plt.rcParams.keys())
```

```python
dist, eff, excl = np.genfromtxt('exclusions_halo2.txt',unpack=True,skip_header=1)
```

### just plot raw data 

```python
plt.scatter(dist,excl)
plt.show()
# Note, you can see a good downward trend until 35 kpc. At that point, stuff is messed up. 
# Some of is because it has trouble creating contours. For example, the exclusion=1 is probably bc no contour was found.
```

### Fix the distance

```python
# CONSTANT DISTANCE
d = 30
eff_ = []
excl_ = []
for i in range(len(dist)):
    if dist[i] == d:
        eff_.append(eff[i])
        excl_.append(excl[i])


plt.scatter(eff_,excl_)
plt.show()
```

### Fix the efficiency

```python
# CONSTANT Efficiency
e = 0.3
dist_ = []
excl_ = []
for i in range(len(eff)):
    if eff[i] == e and dist[i] <= 30:
        dist_.append(dist[i])
        excl_.append(excl[i])

        
plt.scatter(dist_,excl_)
plt.show()
```

### Use good data + interpolation/extrapolation to create full data set

```python
# Good data (mostly)

dist_good = []; eff_good = []; excl_good = []
for i in range(len(dist)):
    if dist[i] <= 30:
        dist_good.append(dist[i])
        eff_good.append(eff[i])
        excl_good.append(excl[i])
```

```python
# Fix the problems

for i in range(len(dist_good)):
    if dist_good[i] == 20 and eff_good[i] == 0.45:
        excl_good[i] = 0.58
    if dist_good[i] == 25 and eff_good[i] == 0.35:
        excl_good[i] = 0.41
    if dist_good[i] == 25 and eff_good[i] == 0.45:
        excl_good[i] = 0.48
    if dist_good[i] == 25 and eff_good[i] == 0.5:
        excl_good[i] = 0.52
    if dist_good[i] == 30 and eff_good[i] == 0.5:
        excl_good[i] = 0.42
    if dist_good[i] == 30 and eff_good[i] == 0.65:
        excl_good[i] = 0.55    
```

```python
# Extrapolation

eff_list = [0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8]
Eff = []; Dist = []; Excl = []
for e in eff_list:
    dist_ = []
    excl_ = []
    
    # get data for this efficiency
    for i in range(len(eff_good)):
        if eff_good[i] == e:
            dist_.append(dist_good[i])
            excl_.append(excl_good[i])
    
    # linear fit and extrapolate
    params = np.polyfit(dist_,excl_,1)
    def f(x):
        return params[0]*x + params[1]
    dist_extend = np.linspace(2,70)
    for d in dist_extend:
        Eff.append(e)
        Dist.append(d)
        Excl.append(f(d))

# clip negative values
Excl = np.clip(Excl,0,1)
```

```python
# interpolate

x = np.linspace(np.min(Dist),np.max(Dist))
y = np.linspace(np.min(Eff),np.max(Eff))
x,y = np.meshgrid(x,y)

z = interpolate.griddata((Dist,Eff),Excl,(x.ravel(),y.ravel()))
        
z = z.reshape(x.shape)
```

```python
# Contour plot        
fig,ax = plt.subplots(1,1,figsize=(8.5,7))
cont = ax.contourf(x,y,z,np.linspace(-1e-5,1,500))
cbar = fig.colorbar(cont,ticks=np.linspace(0,1,6))

ax.set_xlim(5,60)
ax.set_xlabel('Distance (kpc)')
ax.set_ylabel('',fontsize=22)
cbar.ax.set_ylabel('Fraction of models excluded')

fig.savefig('figure_of_merit_halo2.png')

plt.show()
```
