```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
energies = np.genfromtxt('TritonEnergy_SN.txt')
```

```python
figsize = (8,6)
plt.rc('font', family='sans-serif')
plt.rc('axes', labelsize='xx-large')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
text_size = 25
legend_size = 11
```

```python
fig1,ax1 = plt.subplots(1,1,figsize=figsize)

gray1 = patches.Rectangle((0.16,0.1),0.65,1e4,linewidth=1,facecolor='gray',alpha=0.18)
gray2 = patches.Rectangle((0.81,0.1),0.77,1e4,linewidth=1,facecolor='gray',alpha=0.32)
ax1.add_patch(gray1)
ax1.add_patch(gray2)

ax1.text(0.26,0.3,'1n event',color='black',fontsize=text_size)
ax1.text(0.98,0.3,'2n event',color='black',fontsize=text_size)

ax1.hist(energies,bins=120,range=(0,1.8),log=True,histtype='step',color='black')
ax1.set_xlim(0,1.8)
ax1.set_ylim(0.1,1e4)
ax1.set_xlabel('Triton Energy (MeV)')
ax1.set_ylabel('Counts')
plt.show()
```

```python
E, e11, e12, e21, e22 = np.genfromtxt('efficiencies.txt',usecols=(0,3,4,5,6),unpack=True)
```

```python
fig2,ax2 = plt.subplots(1,1,figsize=figsize)

ax2.scatter(E,e11/100.0,color='r',marker='o',label='1n detected as 1n')
ax2.scatter(E,e12/100.0,color='r',marker='s',facecolor='none',label='1n detected as 2n')
ax2.scatter(E,e21/100.0,color='b',marker='o',label='2n detected as 1n')
ax2.scatter(E,e22/100.0,color='b',marker='s',facecolor='none',label='2n detected as 2n')

ax2.set_ylim(-0.01,0.5)

ax2.legend(fontsize=legend_size)

ax2.set_xlabel("Neutron Kinetic Energy (MeV)")
ax2.set_ylabel("Fraction detected")

plt.show()
```
