```python
import matplotlib.pyplot as plt
from modules.sn_models import snowglobes_curves,other_sn_models
```

```python
plt.style.use('jfc') # my custom style sheet
%config InlineBackend.figure_format = 'retina'
```

```python
# sizes for text in plots
exp_size = 22
dist_size = 14
model_size = 14
```

```python
fig1, ax1 = plt.subplots(1,1,figsize=(5.6,4))

snowglobes_curves(ax1,1,5,True)
other_sn_models(ax1,1,5,True)

# Other model labels
ax1.text(85,15,'LS220',color='black',fontsize=model_size)
ax1.text(51,4,'Shen',color='gray',fontsize=model_size)
ax1.text(72,27,'Garching',color='c',fontsize=model_size)
ax1.text(2.5,16,'Basel-Darmstadt',color='forestgreen',fontsize=model_size-2)
ax1.text(51,43,'GVKM',color='m',fontsize=model_size)

# Title
ax1.text(0.2,0.82,'HALO',fontsize=exp_size,ha='center',va='center',transform=ax1.transAxes)
ax1.text(0.2,0.75,'5 kpc',fontsize=dist_size,ha='center',va='center',transform=ax1.transAxes)

# Axes
ax1.set_xlabel('1n events')
ax1.set_ylabel('2n events')

ax1.set_xlim(0,120)
ax1.set_ylim(-2,52)

fig1.savefig('./figures/supernova_models_halo1_5kpc.pdf')

plt.show()
```

```python
fig2, ax2 = plt.subplots(1,1,figsize=(5.6,4))

snowglobes_curves(ax2,1,10,True)
other_sn_models(ax2,1,10,True)

# Title
ax2.text(0.2,0.82,'HALO',fontsize=exp_size,ha='center',va='center',transform=ax2.transAxes)
ax2.text(0.2,0.75,'10 kpc',fontsize=dist_size,ha='center',va='center',transform=ax2.transAxes)

# Axes
ax2.set_xlabel('1n events')
ax2.set_ylabel('2n events')

ax2.set_xlim(0,30)
ax2.set_ylim(-1,13)

fig2.savefig('./figures/supernova_models_halo1_10kpc.pdf')

plt.show()
```

```python
fig3, ax3 = plt.subplots(1,1,figsize=(5.6,4))

snowglobes_curves(ax3,2,10,True)
other_sn_models(ax3,2,10,True)

scale = 1/0.079*1/4

# Other model labels
ax3.text(85*scale,15*scale,'LS220',color='black',fontsize=model_size)
ax3.text(51*scale,4*scale,'Shen',color='gray',fontsize=model_size)
ax3.text(72*scale,27*scale,'Garching',color='c',fontsize=model_size)
ax3.text(2.5*scale,16*scale,'Basel-Darmstadt',color='forestgreen',fontsize=model_size-2)
ax3.text(71*scale,37.5*scale,'GVKM',color='m',fontsize=model_size)

# Title
ax3.text(0.24,0.82,'HALO-1kT',fontsize=exp_size-2,ha='center',va='center',transform=ax3.transAxes)
ax3.text(0.24,0.75,'10 kpc',fontsize=dist_size,ha='center',va='center',transform=ax3.transAxes)

# Axes
ax3.set_xlabel('# 1n events',horizontalalignment='right',x=1.0)
ax3.set_ylabel('# 2n events',horizontalalignment='right', y=1.0)

ax3.set_xlim(-10,scale*120)
ax3.set_ylim(-2,scale*52)

fig3.savefig('./figures/supernova_models_halo2_10kpc.pdf')

plt.show()
```
