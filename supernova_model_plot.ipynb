```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
```

```python
from modules.sn_models import snowglobes_curves,other_sn_models
```

```python
figsize = (8.5,6)
plt.rc('font', family='sans-serif')
plt.rc('axes', labelsize='xx-large')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
exp_size = 30
dist_size = 20
model_size = 18
```

```python
fig, ax = plt.subplots(1,1,figsize=figsize)

snowglobes_curves(ax,1,5,True)
other_sn_models(ax,1,5,True)

# Other model labels
ax.text(85,15,'LS220',color='black',fontsize=model_size)
ax.text(51,4,'Shen',color='gray',fontsize=model_size)
ax.text(72,27,'Garching',color='c',fontsize=model_size)
ax.text(2.5,16,'Basel-Darmstadt',color='forestgreen',fontsize=model_size-2)
ax.text(53,43,'GVKM',color='m',fontsize=model_size)

# Title
ax.text(0.1,0.8,'HALO',fontsize=exp_size,transform=ax.transAxes)
ax.text(0.14,0.74,'5kpc',fontsize=dist_size,transform=ax.transAxes)

# Axes
ax.set_xlabel('# 1n events',horizontalalignment='right',x=1.0)
ax.set_ylabel('# 2n events',horizontalalignment='right', y=1.0)

ax.set_xlim(0,120)
ax.set_ylim(-2,52)

pylab.savefig('./figures/supernova_models_halo1_5kpc.png')

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=figsize)

snowglobes_curves(ax,2,10,True)
other_sn_models(ax,2,10,True)

scale = 1/0.079*1/4

# Other model labels
ax.text(85*scale,15*scale,'LS220',color='black',fontsize=model_size)
ax.text(51*scale,4*scale,'Shen',color='gray',fontsize=model_size)
ax.text(72*scale,27*scale,'Garching',color='c',fontsize=model_size)
ax.text(2.5*scale,16*scale,'Basel-Darmstadt',color='forestgreen',fontsize=model_size-2)
ax.text(53*scale,43*scale,'GVKM',color='m',fontsize=model_size)

# Title
ax.text(0.05,0.8,'HALO-1kT',fontsize=exp_size,transform=ax.transAxes)
ax.text(0.14,0.74,'10kpc',fontsize=dist_size,transform=ax.transAxes)

# Axes
ax.set_xlabel('# 1n events',horizontalalignment='right',x=1.0)
ax.set_ylabel('# 2n events',horizontalalignment='right', y=1.0)

ax.set_xlim(0,scale*120)
ax.set_ylim(-2,scale*52)

pylab.savefig('./figures/supernova_models_halo2_10kpc.png')

plt.show()
```
